# Web Scraping

Web scraping is the process of collecting and processing the large amount of data from web. It extracts the specific data from internet automatically. We can extract data from wikipidea using wikipedia library and if we want to extract data from any other website then we have library named "Beautifulsoup".

```python
! pip install wikipedia

import wikipedia as wiki
```

**Extracting data from wikipedia**

```python
topic = 'Python basic'

wiki.set_lang('en')                        # You can specify any language like french, japanese, german, etc.
para = wiki.page(topic).content            # Extracting the content about topic i.e. python from wikipedia page
print(para)
```

**Cleaning extracted data**

```python
# A regular expression (or RE) specifies a set of strings that matches it.
# The module has functions in this module that let you check if a particular string matches a given regular expression or string.

import re
```

```python
# re.sub(pattern or regular expression, replace with, string or text)

text = re.sub('[^A-Za-z0-9. ]', "", para)          # Here if the character is not capital or small aplhabet, number, dot or space then it will be removed in the para string
text                                               # This text doesn't contain any undesired special character
```

```python
k = re.search("Uses",text)              # search any string or expression in the text
k.group()
```

```python
# finding regular expression in the para

re.findall('==.*==', para)              # this command finds all the strings starting and ending from '=='
```

```python
title_index = html.find("<title>")            # Returns the index of starting of <title>
start_index = title_index + len("<title>")
end_index = html.find("</title>")

title = html[start_index:end_index]
print(title)
```

**Extracting information about the titles, price and image of mmonitors**

```python
titles = []
prices = []
images = []

for info in soup.find_all('div', attrs={'class':"_2kHMtA"}):
  title = info.find('div', attrs={'class':"_4rR01T"})
  titles.append(title.string)

  price = info.find('div', attrs={'class':"_30jeq3 _1_WHN1"})
  prices.append(price.string)

  image = info.find('img', attrs={'class':"_396cs4"})
  images.append(image.get('src'))

  print("Title:", title.string)
  print("Price:",price.string)
  print("Image link:",image.get('src'), "\n\n")
```
